```python
# Incluindo estado civil na base simulada
# Gerar dados com distribuição mais realística (gaussiana truncada)
import pandas as pd
import numpy as np
import random

# Garantir reprodutibilidade
random.seed(42)
np.random.seed(42)

# número de registros simulados
n = 1000
# 1. idade entre 18 e 80 anos
idades = np.random.normal(loc=45, scale=12, size=n).astype(int)
idades = np.clip(idades, 18, 80)

# 2. UF (estados principais)
ufs = np.random.choice(
    ['SP', 'RJ', 'MG', 'RS', 'PR'],
    size=n,
    p=[0.30, 0.20, 0.20, 0.15, 0.15]  # Distribuição mais realista
)

# 3.Tipo de seguro
tipos_de_seguros = np.random.choice(
    ['auto', 'vida', 'residencial'],
    size=n,
    p=[0.6, 0.25, 0.15]  # exemplo de distribuição realista
)

# 4. Renda mensal (simulação aproximada em R$)
renda = np.random.normal(loc=4000, scale=1500, size=n).astype(int)
renda = np.clip(renda, 1200, 20000) # evita valores negativos ou muito extremos

#5. Histórico de sinistro anterior (1:sim, 0:não)
sinistro = np.random.choice([0,1], size=n, p=[0.85, 0.15]) #15% já tiveram

# 6. Estado civil (simulado)
estado_civil = np.random.choice(
    ['solteiro', 'casado', 'divorciado', 'viuvo'],
    size=n,
    p=[0.48, 0.40, 0.06, 0.06]
)

#7. Montar o DataFrame base

df = pd.DataFrame({
    'idade': idades,
    'uf': ufs,
    'tipo_seguro': tipos_de_seguros,
    'renda': renda,
    'sinistro': sinistro,
    'estado_civil': estado_civil
})
df.head()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='estado_civil', data=df, palette='Set1')
plt.title("Distribuição por Estado Civil")
plt.xlabel("Estado civil")
plt.ylabel("Quantidade")
plt.show()
```

```python
# Gerando a variável de risco: baixo, médio e alto
'''
Critérios iniciais de risco:
idade: <25 ou >65 + risco
estado civil: solteiro, separado + risco
renda <2500 + risco
UF RJ e SP + risco
Tipo de seguro: auto + risco que vida/residência
Sinistro anterior: aumenta risco
'''
def classificar_risco(row):
    score = 0
    # Sinistro anterior
    if row['sinistro'] == 1:
        score += 30
    
    # Tipo de seguro
    if row['tipo_seguro'] == 'auto':
        score += 20
    if row['tipo_seguro'] == 'vida':
        score += 10
    if row['tipo_seguro'] == 'residencial':
        score += 5
        
    # Renda
    if row['renda'] < 2500:
        score += 20
    elif row['renda'] < 4000:
        score += 10
    
    # idade
    if row['idade'] < 25 or row['idade'] > 65:
        score += 15
    
    # UF
    if row['uf'] in ['RJ', 'SP']:
        score += 10
    
    # Estado Civil
    if row['estado_civil'] == 'solteiro':
        score += 10
    elif row['estado_civil'] in ['divorsiado', 'viuvo']:
        score += 5
    
    # Classificação
    if score >= 60:
        return 'alto'
    elif score >= 35:
        return 'médio'
    else:
        return 'baixo'
```

```python
df['risco'] = df.apply(classificar_risco, axis=1)
df['risco'].value_counts(normalize=True)
```

# Pré-processamento dos dados
### 1. Objetivo: separar as variáveis de entrada e alvo
- entrada: idade, uf, tipo de seguro, renda, sinistro, estado civil
- alvo: risco

### 2. Codificar variáveis categóricas com OneHotEncoder ou pd.get_dummies
### 3. Dividir dos dados em treino e teste

```python
X = df.drop('risco', axis=1)
y = df['risco']
```

```python
X_encoded = pd.get_dummies(X, columns=['uf','tipo_seguro','estado_civil'], drop_first=True)
```

```python
# Codificar y em classes numéricas
'''
baixo = 0
médio = 1
alto = 2
'''
# Mapeamento manual
mapeamento_risco = {'baixo': 0, 'médio': 1, 'alto': 2}

y_encoded = y.map(mapeamento_risco)

# Dicionário inverso para uso na previsão
classes_risco = {v: k for k, v in mapeamento_risco.items()}
```

```python
# Separar treino e teste: 80% treino e 20% teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded # mantém a proporção das classes
    
)
```

# Treinamento do modelo de Regressão Logística

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

# Escalar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Treinar o modelo sem multi_class
modelo_rl = LogisticRegression(
    solver='lbfgs',
    max_iter=3000,
    random_state=42
)
modelo_rl.fit(X_train_scaled, y_train)
```

```python
# Previsão das classes
y_pred = modelo_rl.predict(X_test_scaled)
```

```python
# Previsão das probabilidades (opcional, útil para dashboards)
y_proba = modelo_rl.predict_proba(X_test_scaled)
```

# Avaliar o desempenho do modelo

```python
from sklearn.metrics import classification_report, confusion_matrix
# Lista de nomes das classes na ordem 0, 1, 2
target_names = [classes_risco[i] for i in sorted(classes_risco)]

print("Relatório de Classificação:")
print(classification_report(y_test, y_pred, target_names=target_names))

print("Matriz de Confusão:")
print(confusion_matrix(y_test, y_pred))
```

# Interpretação dos resultados 
- Acurácia muito boa, 91% dos acertos com base não real ainda
- O modelo aprendeu bem padrões reais, especialmente nas classes baixo e médio
- Pronto para ser salvo e testado, ideal para compartar com outros modelos 
- Classe 'alto' poderia ser reforçada para balancear mais exemplos dela se necessário.
